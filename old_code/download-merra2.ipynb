```python
from requests.auth import HTTPBasicAuth

f = open("prec_evap.txt", "r") # wind


for i, URL in enumerate(f):
    
    URL = URL[:-1]

    if i > 0:
        
        # Set the FILENAME string to the data file name, the LABEL keyword value, or any customized name. 
#         FILENAME = URL[131:170] # wind & raditation
        FILENAME = URL[81:120] # turbulent fluxes & prec evap
        
        import requests
        result = requests.get(URL, auth=HTTPBasicAuth('marceldpl10@gmail.com', 'erie9uMo'))

        result.raise_for_status()
        f = open(FILENAME,'wb')
        f.write(result.content)
        f.close()
        print('contents of URL written to '+FILENAME)
```
