```python
import matplotlib.pyplot as plt
import xarray as xr
import cmocean.cm as cmo
import numpy as np
import my_functions as my
import gsw
from matplotlib.dates import date2num
import matplotlib.dates as mdates

import my_plot_params

years = mdates.YearLocator()   # every year
months = mdates.MonthLocator(interval=1)  # every month
week = mdates.WeekdayLocator(byweekday=mdates.MO, interval=1)
weeks = mdates.WeekdayLocator(byweekday=mdates.MO, interval=3)

yearsFmt = mdates.DateFormatter("%d/%m")

mnthFmt = mdates.DateFormatter("%B")

lightblue = '#5499c7'
blue      = '#21618c'
orange    = '#f39c12'
green     = '#27ae60'
red       = '#cb4335'
```

```python
dat_saz = xr.open_dataset('../data/dat_saz_6H.nc')
dat_pfz = xr.open_dataset('../data/dat_pfz_6H.nc')
dat_miz = xr.open_dataset('../data/dat_miz_6H.nc')

for i, dat in enumerate([dat_saz, dat_pfz, dat_miz]):
    
    dat['tau']=(('time'), np.sqrt(dat['taux']**2 + dat['tauy']**2).data)

dat_saz = dat_saz.resample(time='1D', loffset='12H').mean()
dat_pfz = dat_pfz.resample(time='1D', loffset='12H').mean()
dat_miz = dat_miz.resample(time='1D', loffset='12H').mean()
```

#### Plot the figure 

```python
fig,ax=plt.subplots(3,3, figsize=[15, 11])
fig.subplots_adjust(hspace=0.4)

for i, dat in enumerate([dat_saz, dat_pfz, dat_miz]):
    
    ax[0,i].plot(dat.time, dat.tau, c='k')

    ax[0,i].set_xlim(dat.time[0], dat.time[-1])
    ax[0,i].set_ylim(0, 0.8)
    
    
for i, dat in enumerate([dat_saz, dat_pfz, dat_miz]):
    
    ax[1,i].plot(dat.time, dat.ssr,  c=red)
    ax[1,i].plot(dat.time, dat.str,  c=orange)
    ax[1,i].plot(dat.time, dat.sshf, c=green)
    ax[1,i].plot(dat.time, dat.slhf, c=blue)
    ax[1,i].plot(dat.time, dat.qnet, lw=2.5, c='k')

    ax[1,i].set_xlim(dat.time[0], dat.time[-1])
    ax[1,i].set_ylim(-200, 400)
    
    
for i, dat in enumerate([dat_saz, dat_pfz, dat_miz]):
    
    ax[2,i].plot(dat.time, dat.tp, c=blue)
    ax[2,i].plot(dat.time, dat.e, c=red)

    ax[2,i].set_xlim(dat.time[0], dat.time[-1])
    ax[2,i].set_ylim(-1e-7, 2e-7)
    
for i in range(3):
    for ii in range(3):
        ax[ii,i].xaxis.set_major_locator(weeks) 
        ax[ii,i].xaxis.set_major_formatter(yearsFmt)
        
        if ii<2:
            
            ax[ii,i].set_xticklabels([])
        
ax[0,1].text(dat_pfz.time[8], 0.7, 'Wind stress', c='k')

yloc = 485
ax[1,0].text(dat_saz.time[70], yloc, 'Shortwave', c=red)
ax[1,1].text(dat_pfz.time[0] - np.timedelta64(3,'D'), yloc, 'Longwave', c=orange)
ax[1,1].text(dat_pfz.time[54], yloc, 'Sensible',      c=green)
ax[1,1].text(dat_pfz.time[30], yloc, 'Latent',        c=blue)
ax[1,1].text(dat_pfz.time[83], yloc, 'Net heat flux', c='k')


ax[2,1].text(dat_pfz.time[45], 1.65e-7, 'Evaporation', c=red)
ax[2,1].text(dat_pfz.time[8],  1.65e-7, 'Precipitation', c=blue)

ax[0,0].set_ylabel('N. m$^{-2}$', labelpad=10)
ax[1,0].set_ylabel('W. m$^{-2}$', labelpad=-6)
ax[2,0].set_ylabel('m. s$^{-1}$', labelpad=-1)

props=dict(fontsize=16, fontweight='bold')

ax[0,0].set_title('Subantarctic Zone' , **props)
ax[0,1].set_title('Polar Frontal Zone', **props)
ax[0,2].set_title('Marginal Ice Zone' , **props)

ax[0,0].text(dat_saz.time[2], 0.72, 'a', fontweight='bold', fontsize=14)
ax[0,1].text(dat_pfz.time[2], 0.72, 'b', fontweight='bold', fontsize=14)
ax[0,2].text(dat_miz.time[2], 0.72, 'c', fontweight='bold', fontsize=14)

ax[1,0].text(dat_saz.time[2], 345, 'd', fontweight='bold', fontsize=14)
ax[1,1].text(dat_pfz.time[2], 345, 'e', fontweight='bold', fontsize=14)
ax[1,2].text(dat_miz.time[2], 345, 'f', fontweight='bold', fontsize=14)

ax[2,0].text(dat_saz.time[2], 1.7e-7, 'g', fontweight='bold', fontsize=14)
ax[2,1].text(dat_pfz.time[2], 1.7e-7, 'h', fontweight='bold', fontsize=14)
ax[2,2].text(dat_miz.time[2], 1.7e-7, 'i', fontweight='bold', fontsize=14)


plt.savefig('../figs_submission2/fig4.png', dpi=300)
```

```python
from windrose import WindroseAxes

fig = plt.figure(figsize=[2, 2])

rect=[0.05,0.05,0.9,0.9]

wr=WindroseAxes(fig, rect)

fig.add_axes(wr)
wr.bar(dat_saz['wind_dir'], dat_saz['wind_dir']*0, 
       opening=0.9, 
       normed=True, 
       edgecolor='w', lw=0, color=blue, 
       zorder=10
      )
wr.yaxis.set_ticklabels([])
plt.grid(linewidth=0.1, linestyle='-', zorder=0, color='k')

plt.savefig('../figs_submission2/fig4_saz_windrose.png', dpi=300)
```
