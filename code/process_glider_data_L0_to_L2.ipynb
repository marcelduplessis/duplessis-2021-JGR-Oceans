#### Plot the thermal lag error estimates - upcast - downcasts

Some small thermal lag exists in the salinity data at around 75 m depth. To avoid this issue biasing the data, we will only analyse the data between 10 m and 10 m above the MLD.

```python
rcParams["figure.figsize"] = [8, 8]
rcParams["figure.dpi"] = 100

font = {'family' : 'Arial',
        'weight' : 'ultralight',
        'size'   : 16}

matplotlib.rc('font', **font)

fig, ax = subplots(1, 2)

for i, val in enumerate(np.arange(0, len(dat_saz_L1.time), 2)):
    
    s = dat_saz_L1.salt[:200, val+1]-dat_saz_L1.salt[:200, val]
    t = dat_saz_L1.temp[:200, val+1]-dat_saz_L1.temp[:200, val]
    
    ax[0].plot(s, s.depth, lw=0.5, c='0.75', alpha=0.5, zorder=2)
    ax[1].plot(t, t.depth, lw=0.5, c='0.75', alpha=0.5, zorder=2)
    
[a.grid(zorder=1, c='0.85') for a in ax]
[a.set_ylim(200, 0) for a in ax]

ax[0].set_xlabel('$\Delta$ Salinity')
ax[1].set_xlabel('$\Delta$ Temperature')
ax[0].set_ylabel('Depth (m)')

dwn = np.arange(0, len(dat_saz_L1.time), 2)
up = dwn+1

s = dat_saz_L1.salt[:200, up].values-dat_saz_L1.salt[:200, dwn].values
s = np.nanmean(s, axis=1)

ax[0].plot(s, np.arange(200), c='0.5', lw=3)

t = dat_saz_L1.temp[:200, up].values-dat_saz_L1.temp[:200, dwn].values
t = np.nanmean(t, axis=1)

ax[1].plot(t, np.arange(200), c='0.5', lw=3)
```

Calculate secondary variables - using the `GliderTools` Python package, I have calculated density and MLD from the Level 1 salinity and temperature data

```python
def dens_mld(salt, temp, depth, lon, lat):
    
    dens0 = np.ndarray(shape(salt))
    mld   = np.ndarray(shape(lon))
    
    for i in range(len(lon)):
        
        dens0[:, i] = gt.physics.potential_density(salt[:, i],       temp[:, i], depth, np.tile(lat[i], 1000), np.tile(lon[i], 1000))
        mld[i]      = gt.physics.mixed_layer_depth(np.tile(0, 1000), depth, dens0[:, i], thresh=0.03)
        
    return dens0, mld
```

```python
dens, mld = dens_mld(dat_saz_L1.salt, dat_saz_L1.temp, dat_saz_L1.depth, dat_saz_L1.lon, dat_saz_L1.lat)
```

Assign secondary variables to a new `xarray dataset`  

```python
dat_saz_L2 = dat_saz_L1.copy()

dat_saz_L2=dat_saz_L2.assign(density=(("depth", "time"), dens))
dat_saz_L2=dat_saz_L2.assign_coords(mld=("time", mld))
```

Assigning the metadata for the SAZ Seaglider

```python
from datetime import date

dat_saz_L2.attrs['instrument']              = 'Hydriod Seaglider'
dat_saz_L2.attrs['owner']                   = 'Southern Ocean Carbon and Climate Observatory'
dat_saz_L2.attrs['contact']                 = 'sarahanne.n@gmail.com'
dat_saz_L2.attrs['processing_date']         = str(date.today())
dat_saz_L2.attrs['processing_level']        = '2'
dat_saz_L2.attrs['time_coverate_start']     = str(dat_saz_L2.time[0].values)
dat_saz_L2.attrs['time_coverate_end']       = str(dat_saz_L2.time[-1].values)
dat_saz_L2.attrs['geospatial_vertical_min'] = '0'
dat_saz_L2.attrs['geospatial_vertical_max'] = '1000'
dat_saz_L2.attrs['geospatial_lat_min']      = str(dat_saz_L2.lat.min().values)
dat_saz_L2.attrs['geospatial_lat_max']      = str(dat_saz_L2.lat.max().values)
dat_saz_L2.attrs['geospatial_lon_min']      = str(dat_saz_L2.lon.min().values)
dat_saz_L2.attrs['geospatial_lon_max']      = str(dat_saz_L2.lon.max().values)
```
